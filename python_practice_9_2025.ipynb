11. (3б) Перетворити колонку launched у цілочислену (Int64). Перевірити, чи все пройшло, як очікувалось.

12. (3б) Видалити колонки, де кількість непустих значень становить менше 30% від загальної кількості записів. Можна вручну порахувати і тут тільки видалити. За автоматизацію розрахунку відстоків - додаткові 3 бали.

```python
percent_non_null = df.notnull().mean() * 100

print("Відсоток непустих значень у кожній колонці:")
print(percent_non_null)

df_cleaned = df.loc[:, percent_non_null >= 30]

print("\nФорма після видалення колонок з менше 30% від загальної кількості записів:", df_cleaned.shape)
```

13. (4б) Очистити датафрейм від дублікатів (дублюючихся рядків) та записів, де атака продовжувалась відʼємну або нульову кількість часу (що фізично неможливо). Не забудьте оновити індекс. Скільки рядків було видалено? Як ви це перевірили у коді?

```python
initial_rows = len(df)

df = df.drop_duplicates()

df = df[df["time_end"] > df["time_start"]]

df = df.reset_index(drop=True)

final_rows = len(df)

removed_rows = initial_rows - final_rows
print(f"Видалено рядків: {removed_rows}")
```

15. (9б) Зʼясуйте, який середній відсоток збиття ракет за весь наведений період.

```python
df["destroyed_percent"] = (df["destroyed"] / df["launched"]) * 100

df_valid = df[df["launched"] > 0]

average_percent = df_valid["destroyed_percent"].mean()

print(f"Середній відсоток збиття ракет за весь період: {average_percent:.2f}%")
```

16. (11б) Зʼясуйте 7 найпопулярніших моделей запущених пристроїв за кількістю випущених одиниць за наведений період. Запишіть результат у файл типу json. Скопіюйте зміст файлу та вставте нижче.

```python
import json

top_models_real = df.groupby('model')['launched'].sum().sort_values(ascending=False).head(7)

top_models_dict_real = top_models_real.to_dict()

json_file_path_real = '/content/top_launched_models.json'
with open(json_file_path_real, 'w') as json_file_real:
    json.dump(top_models_dict_real, json_file_real, indent=4)

with open(json_file_path_real, 'r') as file_real:
    json_content_real = file_real.read()

json_content_real, json_file_path_real
```

17. (10б) Зʼясуйте, у який день було найбільше влучань за весь період (незбитих пристроїв). Виведіть час початку атаки та кількість влучань.

```python
df["hits"] = df["launched"] - df["destroyed"]

df_hits = df[df["hits"] > 0]

max_hit_row = df_hits.loc[df_hits["hits"].idxmax()]

print(f"Час початку атаки: {max_hit_row['time_start']}")
print(f"Кількість влучань: {int(max_hit_row['hits'])}")
```

18. (15б) Знайдіть найбільш тривалу та найбільш коротку атаки за кожен представлений місяць. Результат отримайте у такому вигляді:

![текст посилання](https://drive.google.com/uc?id=1gISSxR2mXjTEOgoAk_vQnXJeqoL-eA5b)


Запишіть його у файл типу сsv. Після цього виведіть на екран.

```python
df["attack_duration"] = df["time_end"] - df["time_start"]

df["year"] = df["time_start"].dt.year
df["month"] = df["time_start"].dt.month

min_durations = df.groupby(["year", "month"])["attack_duration"].min().reset_index()
max_durations = df.groupby(["year", "month"])["attack_duration"].max().reset_index()

result_df = pd.concat([min_durations, max_durations], ignore_index=True)

result_df = result_df.sort_values(by=["year", "month"]).reset_index(drop=True)

result_df.to_csv("monthly_attack_durations.csv", index=False)

print("Найбільш короткі та тривалі атаки за кожен місяць:")
print(result_df)
```
